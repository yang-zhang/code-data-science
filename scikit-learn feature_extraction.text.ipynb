```python
import numpy as np
import pandas as pd
import sklearn.datasets
import sklearn.feature_extraction.text
```

```python
categories = ['alt.atheism', 'soc.religion.christian',
              'comp.graphics', 'sci.med']
twenty_train = sklearn.datasets.fetch_20newsgroups(subset='train',
    categories=categories, shuffle=True, random_state=42)

line_ind = 0
line = twenty_train.data[line_ind]
print(line)

count_vect = sklearn.feature_extraction.text.CountVectorizer()

X_train_counts = count_vect.fit_transform(twenty_train.data)

line_countvectorized = X_train_counts.toarray()[line_ind]
```

```python
ind_line_countvectorized_nonzero = np.where(line_countvectorized>0)[0]
```

```python
words = np.array(count_vect.get_feature_names())[ind_line_countvectorized_nonzero]
```

```python
words = [count_vect.get_feature_names()[i] for i in ind_line_countvectorized_nonzero[0]]
```

```python
tf_transformer = sklearn.feature_extraction.text.TfidfTransformer(use_idf=False)
X_train_tf = tf_transformer.fit_transform(X_train_counts)

line_tf = X_train_tf.toarray()[line_ind]

tfidf_transformer = sklearn.feature_extraction.text.TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

line_tfidf = X_train_tfidf.toarray()[line_ind]
```

```python
df_text_features = pd.DataFrame({
    'ind': ind_line_countvectorized_nonzero,
    'word': words,
    'count': line_countvectorized[ind_line_countvectorized_nonzero],
    'tf': line_tf[ind_line_countvectorized_nonzero],
    'tfidf': line_tfidf[ind_line_countvectorized_nonzero],
})
```
