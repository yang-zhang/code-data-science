```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sklearn.model_selection
import sklearn.datasets
import sklearn.ensemble
import xgboost as xgb
```

```python
data = sklearn.datasets.make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=4,
    n_redundant=2,
    n_repeated=1,
    n_classes=2)
```

```python
X = data[0]
y = data[1]
```

```python
mdl = xgb.XGBClassifier()
```

```python
n=10
para_name = 'learning_rate'
para_values = np.logspace(-5, 0, num=n)

train_scores, valid_scores = sklearn.model_selection.validation_curve(
    mdl, X, y, param_name=para_name, param_range=para_values, scoring='neg_log_loss')

plt.plot(np.mean(train_scores, axis=1))
plt.plot(np.mean(valid_scores, axis=1))
plt.xticks(range(n), para_values, rotation='vertical')
```
